# Label Encoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoding = LabelEncoder()

data_set["Area"] = label_encoding.fit_transform(data_set["Area"])
```

```python
data_set.head()
```

# Using for loop

```python
data_set2 = data_set
```

```python
for column in data_set2.columns:
    if data_set2[column].dtype == np.number:
        continue
    else:
        data_set2[column] = label_encoding.fit_transform(data_set2[column])
    
```

```python
data_set2.head()
```

```python
from pandas.core.dtypes.common import is_numeric_dtype
```

```python
for column in data_set2.columns:
    if is_numeric_dtype(data_set2[column]):
        continue 
    else:
        data_set2[column] = label_encoding.fit_transform(data_set2[column])
```

```python
data_set2.tail()
```

```python
data_set4 = pd.read_csv("supershop.csv")
data_set4.head()

columns = ["Area"]

for col in columns :
    one = pd.get_dummies(data_set4[col], drop_first = True)
    
    data_set4 = pd.concat([data_set4,one],axis=1).drop(col,axis=1)
```

# Ordinal Encoder

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
data_set5 = pd.read_csv("supershop.csv")
```

```python
city = ["Dhaka","Rangpur","Ctg"]
ordinal = OrdinalEncoder(categories = [city])
encoded_value = ordinal.fit_transform(data_set5[["Area"]])
```

```python
#encoded_value
```

```python
encoded_area = pd.DataFrame(encoded_value,columns = ["Encoded"])
```
